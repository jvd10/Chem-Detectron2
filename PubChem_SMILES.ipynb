# Filter duplicates and generate statistics

```python
import csv
import statistics as stat

filtered_smiles = set()
total_unfiltered = 0
total_filtered = 0
lens_unfiltered = []
lens_filtered = []
with open('CID-SMILES', 'r') as f:
  cr = csv.reader(f, delimiter='\t')
  for row in cr:
    total_unfiltered += 1
    sm = row[1]
    lens_unfiltered.append(len(sm))
    if sm not in filtered_smiles:
      total_filtered += 1
      filtered_smiles.add(sm)
      lens_filtered.append(len(sm))

```

```python
import numpy as np
print('Unfiltered:')
print(f'Size: {total_unfiltered}')
```

```python
print('Filtered:')
print(f'Size: {total_filtered}')
print(f'Mean: {np.mean(lens_filtered)}')
print(f'Median: {np.median(lens_filtered)}')
print(f'Percentiles: {np.quantile(lens_filtered, [0.0, 0.25, 0.5, 0.75, 1.0])}')
```

# Sample subset from filtered and generate deepsmiles; save dataset as file

```python
import random
import deepsmiles
random.seed(60201)
converter = deepsmiles.Converter(rings=True, branches=True)
#encoded = converter.encode("c1cccc(C(=O)Cl)c1")
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import Draw
from xml.dom import minidom
from scipy.spatial.ckdtree import cKDTree
import sys
```

```python
def _get_unique_atom_smiles_and_rarity(smiles):
    mol = Chem.MolFromSmiles(smiles)
    assert mol, f'INVALID SMILES STRING: {smiles}'
    dm = Draw.PrepareMolForDrawing(mol)
```

```python
k = 30000
filtered_sample = None
still_needed = k
final_sample = []
while still_needed > 0:
  filtered_sample = random.sample(filtered_smiles, still_needed)
  still_needed = 0
  for s in tqdm(filtered_sample):
    try:
      _get_unique_atom_smiles_and_rarity(s)
      final_sample.append(s)
    except:
      still_needed += 1
      print(f'Found {still_needed} samples as invalid.')

filtered_sample = final_sample
filtered_sample_lens = [len(s) for s in filtered_sample]

assert len(filtered_sample) == k, f'Actually sampled {len(filtered_sample)}.'
```

```python
print(f'Filtered {k}-Sample:')
print(f'Mean: {np.mean(filtered_sample_lens)}')
print(f'Median: {np.median(filtered_sample_lens)}')
print(f'Percentiles: {np.quantile(filtered_sample_lens, [0.0, 0.25, 0.5, 0.75, 1.0])}')
```

## Save file with image name, smiles and image name, deepsmiles

Use this block for generating training data .csv file for TRAINING Chem-Detectron2

```python
f1n = f'pubchem_smiles_{k}.csv'
f2n = f'pubchem_deepsmiles_{k}.csv'
with open(f1n, 'w') as fw1, open(f2n, 'w') as fw2:
  cw1 = csv.writer(fw1)
  cw2 = csv.writer(fw2)
  headers = ['file_name', 'SMILES']
  cw1.writerow(headers)
  cw2.writerow(headers)
  for i, smil in enumerate(filtered_sample):
    img_name = f'train_{i}.png'
    deepsmiles_str = converter.encode(smil)
    cw1.writerow([img_name, smil])
    cw2.writerow([img_name, deepsmiles_str])
print(len(filtered_sample))
```

Use this block for generating images used to generating features from Chem-Detectron2 for Chem-OSCAR's inputs

```python
import json
smiles_json = {}
for i, smil in enumerate(filtered_sample):
  img_name = f'test_{i}.png'
  smiles_json[img_name] = smil

fj_name = f"smiles_json_{k}.json"
with open(fj_name, 'w') as fj:
  json.dump(smiles_json, fj)
```
